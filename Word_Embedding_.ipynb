```python
z = open("dataset.txt")
y = z.read()
print(y)
z.close()
```

```python
sent = list(y.split("\n"))
```

```python
sent
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=10000
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)

#this gives index to each and every word
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
#since the words are not of equal length we user padding technique i.e pre here that adds zeros to the front to make the sentences of equal length
```

```python
dim = 15
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse') #mean squared error(performance metrics)
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

Process of performing word embedding
1. get dataset
2. One hot representation - index to each word
3. Embedding layers to form embedding matrix
4. Embedding matrix, vocb_size = 10000, dim = 10
